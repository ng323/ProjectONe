```python
import pandas as pd

file = "Malaria.csv"

malaria_file = pd.read_csv(file, encoding = 'latin-1')
malaria_file.head()

```

```python
malaria_file.drop(['Unnamed: 2'], axis=1, inplace = True)

malaria_file.head()
```

```python
clean_malaria_file = malaria_file.dropna(how='any')
clean_malaria_file.head()
```

```python
df_malaria = pd.DataFrame(clean_malaria_file)
df_malaria.head()
```

```python
clean_malaria_file.to_csv("clean_malaria_file.csv", index=False, header=True)
```
